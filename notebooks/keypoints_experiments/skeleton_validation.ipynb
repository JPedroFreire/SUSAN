```python
import cv2
import numpy as np
import tensorflow as tf
import pandas as pd
import os

from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as preprocess_mobilenet
```

```python
from distutils.dir_util import copy_tree

copy_tree('/content/drive/MyDrive/skeletons_drawn_dataset/', '/content/Skeletons_drawn_dataset/')
```

```python
def get_video_files(folder_path):
    video_files = []

    for root, dirs, files in os.walk(folder_path):
        subfolder_list = root.split(os.path.sep)
        v_nv = subfolder_list[-2]
        c1_c2 = subfolder_list[-1]

        video_files.extend([(v_nv, c1_c2, file) for file in files if file.lower().endswith(('.mp4', '.avi', '.mkv'))])

    return video_files

folder_path = '/content/Skeletons_drawn_dataset/'

video_files_list = get_video_files(folder_path)


for v_nv, c1_c2, video_file in video_files_list:
    print(f"{v_nv}/{c1_c2}/{video_file}")
```

```python
def predict_violence(img_array):
  img_array = tf.expand_dims(img_array, 0)
  preprocess = preprocess_mobilenet(img_array)
  prediction = model_mobilenet.predict(preprocess, verbose=0)

  return prediction
```

```python
model_mobilenet = tf.keras.models.load_model('skeleton_mobilenet.keras')
```

```python
directory_root = '/content/Skeletons_drawn_dataset/'

df_violence_results = pd.DataFrame(columns=['video_name','label', 'camera', 'violence_detected'])

predictions_results = []

for label, cam, vid in video_files_list:
  video_cap = cv2.VideoCapture(directory_root+f'{label}/{cam}/{vid}')

  violence_count = 0

  while video_cap.isOpened():
      ret, frame = video_cap.read()

      if not ret:
        print("Stream end. Exiting.")
        break

      resized = tf.image.resize(frame, (120,160), tf.image.ResizeMethod.NEAREST_NEIGHBOR)
      resized_cast = tf.cast(resized, tf.float32)
      prediction_violence = predict_violence(resized_cast)

      if(prediction_violence[0][1] > prediction_violence[0][0]):
        violence_count += 1

  violence_detected = violence_count > 10

  predictions_results.append(pd.Series([vid, label, cam[-1], violence_detected], index=df_violence_results.columns))

  video_cap.release()
```

```python
result_dataframe = pd.concat(predictions_results,axis=1).T
```

```python
def metrics_detected(df):
  #true positive
  tp = df[(df['violence_detected'] == 1) & (df['label'] == 'violent')].shape[0]

  #true negative
  tn = df[(df['violence_detected'] == 0) & (df['label'] == 'non-violent')].shape[0]

  #false positive
  fp = df[(df['violence_detected'] == 1) & (df['label'] == 'non-violent')].shape[0]

  #false negative
  fn = df[(df['violence_detected'] == 0) & (df['label'] == 'violent')].shape[0]

  accuracy = (tp + tn)/(tp + tn + fp + fn)
  precision = tp/(tp + fp)
  recall = tp/(tp + fn)
  f1_score = (2 * precision * recall)/(precision + recall)
  total = df.shape[0]

  result = {
      'accuracy': accuracy,
      'precision': precision,
      'recall': recall,
      'f1_score': f1_score,
  }

  print(f"True positive: {tp}")
  print(f"True negative: {tn}")
  print(f"False positive: {fp}")
  print(f"False negative: {fn}")
  print(f"Total: {total}")
  print(f"Acurcia: {accuracy}")
  print(f"F1 Score: {f1_score}")
  print(f"Precision: {precision}")
  print(f"Recall: {recall}")
```
